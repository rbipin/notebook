```python
import boto3
import urllib3
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)
import math;

queueUrl = 'https://sqs.us-east-2.amazonaws.com/776333589406/prod-210418-mc-amp-datasync-dead-letter.fifo'
#queueUrl = 'https://sqs.us-east-2.amazonaws.com/842981125669/beta-210418-mc-amp-datasync-dead-letter.fifo'
sqs = boto3.client('sqs', verify=False)
queueAttributes = sqs.get_queue_attributes(QueueUrl=queueUrl, 
                                           AttributeNames=['All'])
numberOfMessagesAvailable = queueAttributes['Attributes']['ApproximateNumberOfMessages']
numberOfMessagesNotVisible = queueAttributes['Attributes']['ApproximateNumberOfMessagesNotVisible']
totalNumberOfMessages = int(numberOfMessagesAvailable) + int(numberOfMessagesNotVisible)
count = math.ceil(int(numberOfMessagesAvailable) / 10)
data_list = []
print(f'Total Number Of Messages {numberOfMessagesAvailable}')

for i in range(count):
  queue = sqs.receive_message(QueueUrl = queueUrl,
                            AttributeNames=['All'], MaxNumberOfMessages=10)
  messages = queue.get('Messages')
  if (messages == None):
    continue
  for message in messages:
    if (message == None):
      continue
    data_list.append(message.get('Body'))

print(f'Messages Received Count: {len(data_list)}')
for item in data_list:
  print(item)
```
